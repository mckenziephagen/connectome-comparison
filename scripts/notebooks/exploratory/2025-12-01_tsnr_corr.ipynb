```python
def tsnr(data,affine,file_name):
    ###Thanks past ariel
    mean_d = np.mean(data,axis=-1)
    std_d = np.std(data,axis=-1)
    tsnr = mean_d/std_d
    tsnr[np.where(np.isinf(tsnr))] = np.nan
    mean_tsnr = np.nanmean(np.ravel(tsnr))
    tsnr_image = nb.Nifti1Image(tsnr,affine)
    nb.save(tsnr_image,file_name)
    savemat(file_name.split('.')[0],{'tsnr':tsnr})
    return mean_tsnr
```
