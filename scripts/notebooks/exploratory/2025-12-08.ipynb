```python
from glob import glob
import os.path as op
import numpy as np
import matplotlib.pyplot as plt
import nibabel as nb
```

```python
import nilearn
import nilearn.datasets
from nilearn.maskers import NiftiLabelsMasker
```

```python
sub_path = '/pscratch/sd/m/mphagen/human-connectome-project-openaccess/HCP1200/derivatives/timeseries/MSMAll_FIX/sub-793465' 
```

```python
ses_1_msmall_file = glob(op.join(sub_path, '*ses-1*NORMED*.nii') ) 
ses_2_file = glob(op.join(sub_path, '*ses-2*.nii') ) 
```

```python
ses_1_msmall_nii = nb.load(ses_2_file[0]).get_fdata() 
```

```python
import seaborn as sns
```

```python
fs=12
```

```python
bbox_params = {'facecolor': 'lightblue', 'alpha': 0.7, 'pad': 4}
n_trs = 200
l_w = 1
ax1 = plt.subplot(411)
ax1.plot(ses_1_msmall_nii[:n_trs,30], lw=l_w)

ax2 = plt.subplot(412, sharex=ax1)
ax2.plot(ses_1_msmall_nii[:n_trs,21], lw=l_w)
ax3 = plt.subplot(413, sharex=ax1)
ax3.plot(ses_1_msmall_nii[:n_trs,41], lw=l_w)

ax4 = plt.subplot(414, sharex=ax1)
ax4.plot(ses_1_msmall_nii[:n_trs,81], lw=l_w)

for axii in [ax1, ax2, ax3, ax4]: 
    axii.set_xticklabels([])
    axii.set_yticklabels([]) 
    axii.set_xticks([])
    axii.set_yticks([])
    axii.patch.set_facecolor('none')

ax1.set_ylabel(r'$ROI_1$', rotation=0, bbox=bbox_params, fontsize=fs,  fontname="Liberation Serif")
ax2.set_ylabel(r'$ROI_2$', rotation=0, bbox=bbox_params, fontsize=fs) 
ax3.set_ylabel(r'$ROI_3$', rotation=0, bbox=bbox_params, fontsize=fs) 
ax4.set_ylabel(r'$ROI_4$', rotation=0, bbox=bbox_params, fontsize=fs) 

ax1.yaxis.set_label_coords(.1, .8, transform=None)
ax2.yaxis.set_label_coords(.1, .8, transform=None)
ax3.yaxis.set_label_coords(.1, .8, transform=None)
ax4.yaxis.set_label_coords(.1, .8, transform=None)

# ax1.set_label_coords(0, 0, transform=None)
plt.subplots_adjust(hspace=-.3, wspace=.1)
# plt.legend(frameon=False)
sns.despine(left=True, bottom=True, right=True)
plt.savefig('minproc_timeseries.svg', transparent=True, format="svg")) 
```
